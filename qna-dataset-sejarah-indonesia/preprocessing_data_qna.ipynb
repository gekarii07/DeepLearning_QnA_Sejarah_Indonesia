Import Dataset Sejarah Indonesia

```python
import pandas as pd
import random
import json
import torch
import shutil
import os
```

```python
df = pd.read_csv('/content/data-DL - GABUNGAN 6 DES 2024.csv')
df
```

Lowering

```python
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
df
```

Menghapus Spasi Diawal dan Akhir Kalimat

```python
df['question'] = df['question'].str.strip()
df['context'] = df['context'].str.strip()
df['answer'] = df['answer'].str.strip()
df
```

Mengecek Data Answer dengan Context dan Mengubah Menjadi Format JSON

```python
# List untuk menampung hasil akhir dalam format JSON
output_data = []
invalid_data = []

# Loop melalui setiap baris di DataFrame
for idx, row in df.iterrows():
    context = row['context']
    question = row['question']
    answer_text = row['answer']

    # Cari posisi `answer_start` dalam `context`
    answer_start = context.find(answer_text)

    # Tentukan apakah answer ditemukan dalam context
    is_impossible = answer_start == -1

    if is_impossible:
        # Simpan data yang tidak valid ke dalam list invalid_data
        invalid_data.append({
            "id": f"{idx:05}",
            "question": question,
            "context": context,
            "answer": answer_text
        })
        continue  # Lewati penyimpanan ke output_data jika data tidak valid

    answers = []

    # Jika answer ditemukan dalam context, tambahkan dalam format JSON
    if not is_impossible:
        answers.append({
            "text": answer_text,
            "answer_start": answer_start
        })

    # Buat struktur qas untuk pertanyaan ini
    qas = {
        "id": f"{idx:05}",  # ID unik untuk setiap pertanyaan
        "is_impossible": is_impossible,
        "question": question,
        "answers": answers
    }

    # Cek apakah context sudah ada dalam output_data
    context_exists = False
    for entry in output_data:
        if entry["context"] == context:
            entry["qas"].append(qas)
            context_exists = True
            break

    # Jika context baru, tambahkan sebagai entry baru
    if not context_exists:
        output_data.append({
            "context": context,
            "qas": [qas]
        })

# Tampilkan data yang tidak valid
if invalid_data:
    print("Data dengan answer yang tidak ditemukan dalam context:")
    for item in invalid_data:
        print(json.dumps(item, indent=4))
else:
    print("Semua answer ditemukan dalam context.")

# Data tidak disimpan jika ada entry tidak valid
if not invalid_data:
    # Simpan ke file JSON
    with open('/content/deep_learning_data.json', 'w') as f:
        json.dump(output_data, f, indent=4)

    print("Data berhasil diubah ke format JSON dan disimpan sebagai 'deep_learning_data.json'")
```

Split Dataset (70:15:15)

```python
# Load data dari file JSON asli
with open('/content/deep_learning_data.json', 'r') as f:
    data = json.load(f)

# Acak data untuk memastikan pemisahan acak
random.shuffle(data)

# Tentukan indeks pemisah berdasarkan perbandingan 70:15:15
train_split_index = int(0.7 * len(data))
val_split_index = int(0.85 * len(data))

# Bagi data menjadi data latih (train), validasi (val), dan uji (test)
train_data = data[:train_split_index]
val_data = data[train_split_index:val_split_index]
test_data = data[val_split_index:]

# Simpan data latih ke file JSON train_sejarah.json
with open('/content/train_sejarah.json', 'w') as f:
    json.dump(train_data, f, indent=4)

# Simpan data validasi ke file JSON val_sejarah.json
with open('/content/val_sejarah.json', 'w') as f:
    json.dump(val_data, f, indent=4)

# Simpan data uji ke file JSON test_sejarah.json
with open('/content/test_sejarah.json', 'w') as f:
    json.dump(test_data, f, indent=4)

print("Data berhasil dipisah dan disimpan ke 'train_sejarah.json', 'val_sejarah.json', dan 'test_sejarah.json'")
```

Menampilkan Jumlah Data Train, Validation, dan Test

```python
# Paths to the JSON files
test_path = '/content/test_sejarah.json'
train_path = '/content/train_sejarah.json'
val_path = '/content/val_sejarah.json'

# Function to count the number of questions (qas) in the dataset
def count_qas(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
        return sum(len(entry['qas']) for entry in data)

# Count the number of questions in each dataset
test_count = count_qas(test_path)
train_count = count_qas(train_path)
val_count = count_qas(val_path)

print(f"Jumlah data di test file: {test_count}")
print(f"Jumlah data di train file: {train_count}")
print(f"Jumlah data di validation file: {val_count}")
```
